```python
#Contain data about 5000(4803) movies(movies dataset)
df = pd.read_csv("tmdb_5000_credits.csv")
df1 = pd.read_csv("tmdb_5000_movies.csv")
print(df.shape)
print(df1.shape)
```

```python
#For ease joining the data frames df and df1 on id column in df1 and movies_id column in df(as they are same and just the movies id)
df2 = df.rename(index = str,columns = {"movie_id":"id"})
df_combined = df1.merge(df2,on="id")
df_combined.columns
```

```python
df_combined.head(5)
```

```python
#Now basically looking at all the columns there are certain columns which will appear of really of no use so better is to drop that columns
#This is a must step as always there are lot of columns in dataset which are less relevant to our problems
df_combined_changed = df_combined.drop(["homepage","status","title_x","title_y","production_countries"],axis = 1)
```

```python
#Info of the dataframe
df_combined_changed.info()
```

```python
#Using weighted average for each movie average rating
#Calculating the components using dataframe for weighted average
v=df_combined_changed["vote_count"]
r=df_combined_changed["vote_average"]
c=df_combined_changed["vote_average"].mean()
m=df_combined_changed["vote_count"].quantile(0.70)
```

```python
#The formula to calculate wighted average is below on right side of equal(it is taken from http://trailerpark.weekly.com)
df_combined_changed["weighted_average"] = ((r*v)+(c*m))/(v+m)
```

```python
#It will be basically a panda series type
df_combined_changed["weighted_average"].head(5)
```

Basically what this weighted_average column is telling is that say in case if a person has wathched the movie The shawshank redemption then the next movie which can be recommended to him is The godfather and then fight club as they have weighted average close to the The shawshank redemption(This can come under content and as well as collaborative filtering because the weighted average calculated for each movie contains both the properties of user behaviour as well as the movie content)-This is collectively on all the user for movies.

```python
df_combined_changed[["original_title","vote_count","vote_average","weighted_average","popularity"]].sort_values(by = "weighted_average",ascending=False).head(5)
```

```python
df_x = df_combined_changed[["original_title","vote_count","vote_average","weighted_average","popularity"]].sort_values(by = "weighted_average",ascending=False)
```

```python
import seaborn as sns
sns.pairplot(df_x)
#Visualizing the variations of df_x features using pairplot
#A lot of inference can be developed out from here(about the weighted average and how does it get affected by various features)
```

```python
#plotting the popularity against weighted_average and it seems that the fashion in which movies are rated(including the count)does not much depends on the popularity of movie
#But we can assume that popularity should also plays a role in the movie recommendation(as they are the movies every one like to watch) 
#So combining both the weighted average concept and the popularity to recommend movies(both given a equal or unequal fashioned weightage)
sns.jointplot(x="weighted_average", y="popularity", data=df_x)
```

```python
#As our popularity and weighted average are on different scales so sclaing them both to same scale
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_combined_changed[["weighted_average","popularity"]])
df_scaled_final = pd.DataFrame(df_scaled,columns=["weighted_average","popularity"])
```

```python
df_combined_changed[["scaled_weighted_average","scaled_popularity"]] = df_scaled_final
```

```python
df_combined_changed.columns
```

```python
#Now giving importance to both the scaled_weighted_average and scaled_popularity
#Giving 30% popularity to scaled_popularity and remaining 70% to scaled_weighted_average
df_combined_changed["final_score"] = df_combined_changed["scaled_weighted_average"]*0.7+df_combined_changed["scaled_popularity"]*0.3
#Now taking ("original_title","vote_count","vote_average","weighted_average","popularity","final_score") from our dataset
df_xx = df_combined_changed[["original_title","vote_count","vote_average","scaled_weighted_average","scaled_popularity","final_score"]]
```

Now basically the score value help in recommendation of movies to users(is similar way we used weighted average)
